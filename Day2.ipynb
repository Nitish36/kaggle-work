```python
#data for the graph
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("Mumbai.csv")
df[0:10]
```

```python
#for normal graph
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.plot(df["DAY"][0:10],df["TEMPERATURE"][0:10],"g*--",ms=10,mfc="blue",mec="red")
plt.legend()
plt.show()
```

```python
#for bar chart
plt.bar(df["DAY"][0:20],df["TEMPERATURE"][0:20])
plt.show()
```

```python
a = df.loc[df['TEMPERATURE']==70.2].count()[0]
b = df.loc[df['TEMPERATURE']<70.2].count()[0]
c = df.loc[df['TEMPERATURE']>70.2].count()[0]

d = [a,b,c]

l = ['TEMPERATURE=70.2','TEMPERATURE<70.2','TEMPERATURE>70.2']
c = ['#712B75','#C74B50','#D49B54']

plt.pie(d,labels=l,autopct='%.2f',colors=c)
plt.show()
```
