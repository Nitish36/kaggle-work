```python
import pandas as pd
df=pd.read_csv("diabetes.csv")
df[0:3]
```

```python
#plt.figure(figsize=(3,5),dpi=300)#to diplay data with dimension pixels
plt.xlabel("Glucose")#Label X 
plt.ylabel("Insulin")#label y
plt.title("Diabaetes Data")#Title
#plt.scatter(df['Glucose'][50:65],df["Insulin"][50:65])#Scatter the data points
plt.plot(df['Glucose'][50:65],df["Insulin"][50:65],'g*--',ms=10,mfc="red",mec="yellow",label="Insulin")
plt.plot(df['Glucose'][50:65],df["Age"][50:65],'rs-.',ms=12,mfc="green",mec="blue",label="Age")
#[color,marker,linestyle]-ls=--,-.
plt.legend()#Display label
plt.show()
#mfc=markerfacecolor
#mec=markeredgecolor,ls-linestyle,
```

```python
df=pd.read_csv("diabetes.csv")
x=df.loc[df["Age"]==50].count()[0]
y=df.loc[df["Age"]<50].count()[0]
z=df.loc[df["Age"]>50].count()[0]
a=["Age==50","Age<50","Age>50"]
d=[0,0,0]
c=["green","red","orange"]
plt.pie([x,y,z],labels=a,autopct="%1.2f%%",colors=c,explode=d)
plt.show()
```
