```python
df1=df.drop(['car_ID','symboling','aspiration','doornumber','carbody','drivewheel'],axis=1)
```

```python
df1['CarName'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
```

```python
car=l.fit_transform(df1['CarName'])
```

```python
fuel=l.fit_transform(df1['fueltype'])
eng=l.fit_transform(df1['enginelocation'])
```

```python
df1['CarBrand']=car
df1["FuelType"]=fuel
df1['Engine']=eng
```

```python
df2=df1.drop(['CarName','fueltype','enginelocation','enginetype','fuelsystem','cylindernumber'],axis=1)
```

```python
X=df2.drop('price',axis=1)
```

```python
y=df2['price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.40)
#60% train data and 40% test data
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
#predict=mx+c
price=reg.predict(X_test)
```

```python
price
```

```python
from sklearn.metrics import mean_squared_error,r2_score
r2_score(y_test,price)
mean_squared_error(y_test,price)
```
